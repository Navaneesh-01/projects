```python
g = df["win_by_runs"]
batting_first =df[g!=0]
```

```python
batting_first.head()
```

```python
v = batting_first["winner"].value_counts()[0:5]
m = batting_first["winner"].value_counts()[0:5].keys()
```

```python
plt.figure(figsize=(15,5))
plt.bar(m,v,color=["b","y","silver","k","r"],width=0.3)
plt.title("Most Wins By Batting First")
plt.xlabel("Teams")
plt.ylabel("Count")
plt.show()
```

```python
label = batting_first["winner"].value_counts()[0:5].keys()
plt.pie(v,labels=label,explode=[0,0,0,0,0.1],shadow=True,autopct="%0.2f%%")
plt.title("Most Wins By Batting First")
plt.show()
```
